```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

stock_num = input('股票代碼：')
stock_num = stock_num.split('/')
start = input('起始年-月-日：')
end = input('結束年-月-日：')
date_start = datetime.strptime(start, '%Y-%m-%d')
date_end = datetime.strptime(end, '%Y-%m-%d')
df_dict = {}
for i in stock_num:
    df_dict[i] = yf.download(i, start = start, end = end)
for symbol, df in df_dict.items():
    df.to_csv('/content/drive/MyDrive/Python Data Analysis & ML/DATA/{}_{}_{}.csv'.format(symbol.split('.')[0], datetime.strftime(date_start, '%Y%m%d'), datetime.strftime(date_end, '%Y%m%d')), encoding='utf-8', index=True)
```

```python
for i in df_dict.values():
  print(i)
```

```python
for i in df_dict:
    plt.plot(df_dict[i]['Adj Close'])
plt.legend(df_dict.keys())
plt.ylabel('Close')
plt.xticks(rotation=90)
```
