```python
# æƒ…æ„Ÿåˆ†æ•¸
import nltk
import jieba
from snownlp import SnowNLP
from nltk.corpus import stopwords
import opencc

# ä¸‹è¼‰åœç”¨è©èªæ–™åº«
nltk.download('stopwords')
# è¼‰å…¥ä¸­æ–‡åœç”¨è©
stop_words = list(stopwords.words('chinese'))
# è½‰æ›ä¸­æ–‡åœç”¨è©åº«ç‚ºç¹é«”ä¸­æ–‡
converter = opencc.OpenCC('s2t.json')
stop_words_zh = [converter.convert(word) for word in stop_words]
# æ‰‹å‹•åŠ å…¥åœç”¨è©
noise = ['ï¼Œ']
stop_words_zh += noise

# å®šç¾©ä¸€å€‹å‡½å¼ï¼Œç”¨æ–¼è¨ˆç®—æƒ…ç·’å¥½æ„Ÿåº¦å¾—åˆ†
def sentiment_score(text):
    # åˆ†è©
    seg_list = jieba.cut(text)
    # å»é™¤åœç”¨è©
    words = [word for word in seg_list if word not in stop_words_zh]
    # åˆä½µæˆä¸€å€‹å­—ç¬¦ä¸²ï¼Œå–å‰å…©ç™¾å€‹å­—åšç‚ºæ‘˜è¦
    text = " ".join(words)[:200]
    # ä½¿ç”¨snownlpé€²è¡Œæƒ…ç·’åˆ†æï¼Œå¾—åˆ°æ­£å‘æƒ…ç·’å€¼
    s = SnowNLP(text)
    return s.sentiments
```

```python
# è¨ˆç®—contentçš„æƒ…æ„Ÿåˆ†æ•¸ï¼Œä¸¦å°‡scoreå­˜å„²æˆä¸€å€‹list
score_list = []
for content in data["content"]:
    score = sentiment_score(content)
    score_list.append(score)
print(score_list)
```

```python
# æ–°å¢ä¸€æ¬„sentiment_scoreä¸¦å°‡æ•¸å­—åŠ å…¥
data = data.assign(sentiment_score = score_list)
data
```

```python
# å°‡contenté‡è¤‡çš„åˆªé™¤
data = data.drop_duplicates(subset='content')
data
```

```python
# æœ€å¾Œå°‡indexé‡æ–°æ’åº
data = data.reset_index(drop=True)
# åˆªé™¤éŒ¯èª¤è³‡æ–™
data = data[data['time'] != '20230600'].reset_index(drop=True)
data
```
