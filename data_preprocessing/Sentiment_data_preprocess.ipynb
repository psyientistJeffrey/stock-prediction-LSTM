```python
import pandas as pd
import numpy as np
```

```python
# 將多筆 csv 檔合併

data = pd.DataFrame()

# 迴圈處理每個CSV檔案
for i in range(1, 15):  # 先假設有這麼多篇(總共有n篇後面的數字就改成n+1篇)
    filename = f"CMoney{i}.csv"
    df = pd.read_csv(filename)
    
    # 將每個CSV檔案的資料合併到data中
    data = pd.concat([data, df], ignore_index=True)
data
```

```python
# 把topic刪掉
data.drop('topic', axis=1, inplace = True)
data.drop('Unnamed: 0', axis=1, inplace = True)
data
```

```python
# 僅留下tag中有提及"加權指數"的列就好
#pd.set_option('display.max_rows', 10) 
data = data[data['tag'].str.contains("加權指數")]
data
```

```python
data.shape[0] 
```

```python
# drop 掉 time 中有數字以外的列(僅留下含有2023的列)
data = data[data['time'].str.contains("2023")]
data
```

```python
# drop 掉沒有 content 的
data = data[data['content'].notnull()]
data
```

```python
# 將like,donate,comment中的空值轉換成 0
data['like'].fillna(0,inplace=True)
```

```python
# 情感分數
import nltk
import jieba
from snownlp import SnowNLP
from nltk.corpus import stopwords
import opencc

# 下載停用詞語料庫
nltk.download('stopwords')
# 載入中文停用詞
stop_words = list(stopwords.words('chinese'))
# 轉換中文停用詞庫為繁體中文
converter = opencc.OpenCC('s2t.json')
stop_words_zh = [converter.convert(word) for word in stop_words]
# 手動加入停用詞
noise = ['，']
stop_words_zh += noise

# 定義一個函式，用於計算情緒好感度得分
def sentiment_score(text):
    # 分詞
    seg_list = jieba.cut(text)
    # 去除停用詞
    words = [word for word in seg_list if word not in stop_words_zh]
    # 合併成一個字符串，取前兩百個字做為摘要
    text = " ".join(words)[:200]
    # 使用snownlp進行情緒分析，得到正向情緒值
    s = SnowNLP(text)
    return s.sentiments
```

```python
# 計算content的情感分數，並將score存儲成一個list
score_list = []
for content in data["content"]:
    score = sentiment_score(content)
    score_list.append(score)
print(score_list)
```

```python
# 新增一欄sentiment_score並將數字加入
data = data.assign(sentiment_score = score_list)
data
```

```python
# 將content重複的刪除
data = data.drop_duplicates(subset='content')
data
```

```python
# 最後將index重新排序
data = data.reset_index(drop=True)
# 刪除錯誤資料
data = data[data['time'] != '20230600'].reset_index(drop=True)
data
```
